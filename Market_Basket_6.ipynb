```python
df = pd.read_csv(r"C:\Users\saurabh pawar\Market_Basket_Optimisation.csv", header = None)
df.head()
```

```python
df.shape
```

```python
transaction=[]
for i in range(0,7501):
    transaction.append([str (df.values[i,j]) for j in range(0,20)])
```

```python
rule_list=apriori(transaction,min_support=0.004,min_confidence=0.03,min_lift=3,min_length=2)
results=list(rule_list)
```

```python
for i in results:
    print('\n');
    print(i);
    print('*********');
    print(len(results));
```
